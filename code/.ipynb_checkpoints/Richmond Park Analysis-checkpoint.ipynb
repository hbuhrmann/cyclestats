```python
import thinkstats2
import thinkplot
import numpy as np
```

```python
query = 'LatitudeDegrees >= 51.421213 & LatitudeDegrees <= 51.456435 & LongitudeDegrees >= -0.308058 & LongitudeDegrees <= -0.252830'
richmonddf = cycledf.query(query)

print(f'cycle sample count = {len(cycledf)} and richmond sample count = {len(richmonddf)}')

cycledf.tail()
```

# Probability Mass Functions for level, downhill and uphill

```python
uphill = richmonddf.query('Speedmps < 20 & Speedmps > 0 & Gradient > 0.01')
level = richmonddf.query('Speedmps < 20 & Speedmps > 0 & Gradient <= 0.01 & Gradient >= -0.01')
downhill = richmonddf.query('Speedmps < 20 & Speedmps > 0 & Gradient < -0.01')

uphillpmf=thinkstats2.Pmf(uphill.Speedmps)
levelpmf=thinkstats2.Pmf(level.Speedmps)
downhillpmf = thinkstats2.Pmf(downhill.Speedmps)

thinkplot.Pmf(levelpmf,label='Level')
thinkplot.Config(xlim=[0,20])
thinkplot.show()

thinkplot.Pmf(uphillpmf,label='Uphill')
thinkplot.Config(xlim=[0,20])
thinkplot.show()

thinkplot.Pmf(downhillpmf,label='Downhill')
thinkplot.Config(xlim=[0,20])
thinkplot.show()

print(f'Level:      Mean: {round(levelpmf.Mean(),3)},  Median: {round(levelpmf.Median(),3)}     Std: {round(levelpmf.Std(),3)}' )
print(f'Uphill:     Mean: {round(uphillpmf.Mean(),3)},  Median: {round(uphillpmf.Median(),3)}     Std: {round(uphillpmf.Std(),3)} ' )
print(f'Downhill:   Mean: {round(downhillpmf.Mean(),3)},  Median: {round(downhillpmf.Median(),3)}     Std: {round(downhillpmf.Std(),3)} ' )

```

# Continuous Distribution Function : Level, Uphill, Downhill

```python
uphillcdf=thinkstats2.Cdf(np.round(uphill.Speedmps,1))
levelcdf=thinkstats2.Cdf(np.round(level.Speedmps,1))
downhillcdf=thinkstats2.Cdf(np.round(downhill.Speedmps,1))



thinkplot.Cdf(levelcdf,label='Level')
thinkplot.Cdf(uphillcdf,label='Uphill')
thinkplot.Cdf(downhillcdf,label='Downhill')
thinkplot.Config(xlim=[0,20])
thinkplot.show()

print(f'Level:      5th Perc: {levelcdf.Percentile(5)}      Median: {levelcdf.Percentile(50)}   95th Perc: {levelcdf.Percentile(95)}' )
print(f'Uphill:     5th Perc: {uphillcdf.Percentile(5)}      Median: {uphillcdf.Percentile(50)}   95th Perc: {uphillcdf.Percentile(95)}' )
print(f'Downhill:     5th Perc: {downhillcdf.Percentile(5)}      Median: {downhillcdf.Percentile(50)}   95th Perc: {downhillcdf.Percentile(95)}' )

```

```python
#The distribution of speed looks like a normal distribution, lets compare it against a normal probability plot

mean = richmondcadencedf.Speedmps.mean()
median=richmondcadencedf.Speedmps.median()
std=richmondcadencedf.Speedmps.std()

xs = [-4,4]
fxs,fys = thinkstats2.FitLine(xs,inter=mean,slope=std)

thinkplot.Plot(fxs,fys, color = 'gray', label = 'model')

xs,ys=thinkstats2.NormalProbability(richmondcadencedf.Speedmps)

thinkplot.Plot(xs,ys,label = 'Speedmps')

#And it is definitely a normal distribution
```

```python
#Lets see if the distribution is consistent over all gradients

min_gr = 0.0
max_gr = 0.12
min_gr_qstr = str(min_gr)
max_gr_qstr = str(max_gr)
min_dstr = str(round(min_gr*100,1))+'%'
max_dstr = str(round(max_gr*100,1))+'%'

query = 'Cadence > 0 & Speedmps > 0 & Speedmps <= 20 & Cadence < 130 and Gradient >= '+min_gr_qstr+' and Gradient <= '+max_gr_qstr
upcadf = richmonddf.query(query)

print(f'Gradients between {min_dstr} and {max_dstr}')

print (f'Sample size of Richmond records with cadence and speed {len(upcadf)}')

print(f'Number of rides = {len(upcadf.ActivityId.value_counts())}')

print(f'Mean = {round(upcadf.Speedmps.mean(),3)},   Median={round(upcadf.Speedmps.median(),3)},   Std={round(upcadf.Speedmps.std(),3)}')

#SCATTERPLOT CADENCE AND GRADIENT
thinkplot.scatter(xs = upcadf.Cadence,ys=upcadf.Gradient,alpha=0.01)
thinkplot.config(xlabel='Cadence',ylabel='Gradient',title='Scatterplot : Cadence vs Gradient for sections  between '+min_dstr+' and '+max_dstr)
thinkplot.show()

#SCATTERPLOT SPEED AND GRADIENT
thinkplot.scatter(xs = upcadf.Speedmps,ys=upcadf.Gradient,alpha=0.01)
thinkplot.config(xlabel='Speedmps',ylabel='Gradient',title='Scatterplot : Gradient vs Speed for sections  between '+min_dstr+' and '+max_dstr)
thinkplot.show()


#CDF CADENCE
cadencecdf = thinkstats2.Cdf(upcadf.Cadence,label='Cadence between '+min_dstr+' and '+max_dstr)
thinkplot.Cdf(cadencecdf)
thinkplot.config(title='CDF for cadence between '+min_dstr+' and '+max_dstr,xlabel='Cadence')
thinkplot.Show()

#CDF SPEED
cadencecdf = thinkstats2.Cdf(upcadf.Speedmps,label='Speed on sections  between '+min_dstr+' and '+max_dstr)
thinkplot.Cdf(cadencecdf)
thinkplot.config(title='CDF for speed on sections between '+min_dstr+' and '+max_dstr,xlabel='Speedmps')
thinkplot.Show()

#NORMAL DISTIBRUTION FIT USING THE MEAN AND STD OF THE DATA
mean = upcadf.Speedmps.mean()
median=upcadf.Speedmps.median()
std=upcadf.Speedmps.std()

xs = [-4,4]
fxs,fys = thinkstats2.FitLine(xs,inter=mean,slope=std)

thinkplot.Plot(fxs,fys, color = 'gray', label = 'Model')

xs,ys=thinkstats2.NormalProbability(upcadf.Speedmps)

thinkplot.Plot(xs,ys,label = 'Speedmps')
thinkplot.Config(title = 'Normal distribution fit for speed on sections between '+min_dstr+' and '+max_dstr)
thinkplot.Show()


#AND LASTLY DO A HISTOGRAM
uphist = thinkstats2.Hist(np.round(upcadf.Speedmps,1),label='Speedmps')

#for x in sorted(uphist.Items(), key=lambda kv: kv[0]):
#    print (x)

thinkplot.Hist(uphist)
thinkplot.Config(title='Histogram of speeds for a gradient  between '+min_dstr+' and '+max_dstr)
thinkplot.Show()

```

```python
#The data above looks a bit  weird, I would have assumed a more continuous distribution
#this might be because the dataset covers a very small geographic area

#Lets see what it does with my whole of East Sussex Dataset, starting with a slightly smaller bit


filepath = '/Users/hbuhrmann/PycharmProjects/CycleStats/TestResults/'
filename= 'east_sussex.csv'

cycledf=cycle.getfinalcsvcycledata(filepath,filename)

#Limit the dataset to Hastings/Seven Sisters

query = 'LatitudeDegrees >= 50.731827 & LatitudeDegrees <= 50.960866 & LongitudeDegrees >= 0.067130 & LongitudeDegrees <= 0.745465'
hastingsdf = cycledf.query(query)

print(f'cycle sample count = {len(cycledf)} and hastings sample count = {len(hastingsdf)}')

cycledf.tail()
```

```python
#Lets see if the distribution is consistent over all gradients

hastingsdf = cycledf 

min_gr = 0.05
max_gr = 0.06
min_gr_qstr = str(min_gr)
max_gr_qstr = str(max_gr)
min_dstr = str(round(min_gr*100,1))+'%'
max_dstr = str(round(max_gr*100,1))+'%'
location = 'East Sussex'

query = 'Cadence > 0 & Speedmps > 0 & Speedmps <= 20 & Cadence < 130 and Gradient >= '+min_gr_qstr+' and Gradient <= '+max_gr_qstr
upcadf = hastingsdf.query(query)

print(f'Gradients between {min_dstr} and {max_dstr}')

print (f'Sample size of {location} records with cadence and speed {len(upcadf)}')

print(f'Number of rides = {len(upcadf.ActivityId.value_counts())}')

print(f'Mean = {round(upcadf.Speedmps.mean(),3)},   Median={round(upcadf.Speedmps.median(),3)},   Std={round(upcadf.Speedmps.std(),3)}')

thinkplot.Clf()

minY=upcadf.Gradient.min()
maxY=upcadf.Gradient.max()

#SCATTERPLOT CADENCE AND GRADIENT
thinkplot.scatter(xs = upcadf.Cadence,ys=upcadf.Gradient,alpha=0.01)
thinkplot.config(xlabel='Cadence',ylabel='Gradient',title='Scatterplot : Cadence vs Gradient for sections  between '+min_dstr+' and '+max_dstr)
thinkplot.config(ylim=[minY,maxY])
thinkplot.show()

#SCATTERPLOT SPEED AND GRADIENT
thinkplot.scatter(xs = upcadf.Speedmps,ys=upcadf.Gradient,alpha=0.01)
thinkplot.config(xlabel='Speedmps',ylabel='Gradient',title='Scatterplot : Gradient vs Speed for sections  between '+min_dstr+' and '+max_dstr)
thinkplot.config(ylim=[minY,maxY])
thinkplot.show()

#JITTERY SCATTERPLOT SPEED AND GRADIENT
speed = thinkstats2.Jitter(upcadf.Speedmps,0.25)
gradient = thinkstats2.Jitter(upcadf.Gradient,0.001)
thinkplot.scatter(xs = speed,ys=gradient,alpha=0.01)
thinkplot.config(xlabel='Speedmps',ylabel='Gradient',title='Jittery Scatterplot : Gradient vs Speed for sections  between '+min_dstr+' and '+max_dstr)
thinkplot.config(ylim=[minY,maxY])
thinkplot.show()





#CDF CADENCE
cadencecdf = thinkstats2.Cdf(upcadf.Cadence,label='Cadence between '+min_dstr+' and '+max_dstr)
thinkplot.Cdf(cadencecdf)
thinkplot.config(title='CDF for cadence between '+min_dstr+' and '+max_dstr,xlabel='Cadence')
thinkplot.Show()

#CDF SPEED
cadencecdf = thinkstats2.Cdf(upcadf.Speedmps,label='Speed on sections  between '+min_dstr+' and '+max_dstr)
thinkplot.Cdf(cadencecdf)
thinkplot.config(title='CDF for speed on sections between '+min_dstr+' and '+max_dstr,xlabel='Speedmps')
thinkplot.Show()

#NORMAL DISTIBRUTION FIT USING THE MEAN AND STD OF THE DATA
mean = upcadf.Speedmps.mean()
median=upcadf.Speedmps.median()
std=upcadf.Speedmps.std()

xs = [-4,4]
fxs,fys = thinkstats2.FitLine(xs,inter=mean,slope=std)

thinkplot.Plot(fxs,fys, color = 'gray', label = 'Model')

xs,ys=thinkstats2.NormalProbability(upcadf.Speedmps)

thinkplot.Plot(xs,ys,label = 'Speedmps')
thinkplot.Config(title = 'Normal distribution fit for speed on sections between '+min_dstr+' and '+max_dstr)
thinkplot.Show()


#AND LASTLY DO A HISTOGRAM
uphist = thinkstats2.Hist(np.round(upcadf.Speedmps,1),label='Speedmps')

#for x in sorted(uphist.Items(), key=lambda kv: kv[0]):
#    print (x)

thinkplot.Hist(uphist)
thinkplot.Config(title='Histogram of speeds for a gradient  between '+min_dstr+' and '+max_dstr)
thinkplot.Show()



```
